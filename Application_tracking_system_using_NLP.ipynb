```python
#Data Cleaning
"""
1. urls
2. hashtags
3. mentions
4. special letters
5. punctuations

"""
```

```python
import re
```

```python
def cleanResume(txt):
  cleantext = re.sub('https\S+\s'," ",txt)
  cleantext = re.sub('#\S+\s'," ",cleantext)
  cleantext = re.sub('RT|cc'," ",cleantext)
  cleantext = re.sub('@\S+\s'," ",cleantext)
  cleantext = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""")," ",cleantext)
  return cleantext
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
# Cleaned data
print(df['Resume'][1])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Category'] = le.fit_transform(df['Category'])
```

```python
df.head()
```

```python
df.tail()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf.fit(df['Resume'])
```

```python
requiredtext = tfidf.transform(df['Resume'])
```

```python
print(requiredtext)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(requiredtext, df['Category'],test_size=0.2, random_state = 87 )
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
#Predcition Part

import pickle

pickle.dump(tfidf,open('tfidf.pkl','wb'))
pickle.dump(model,open('model.pkl','wb'))
```

```python
resume = """ working as data analyst / data scientist

 """
x = pickle.load(open('model.pkl','rb'))
cleaned_resume = cleanResume(resume)

input_features = tfidf.transform([cleaned_resume])

prediction_id = x.predict(input_features)[0]
category_mapping = {
               15 : 'Java Developer',
               23 : 'Testing',
               8 : 'DevOps Engineer',
               20 : 'Python Developer',
               24 : 'Web Designing',
               12 : 'HR',
               13 : 'Hadoop',
               3 : 'Blockchain',
               10 : 'ETL Developer',
               18 : 'Operations Manager',
               6 : 'Data Science',
               22 : 'Sales',
               16 : 'Mechanical Engineer',
               1 : 'Arts',
               7 : 'Database',
               2 : 'Business Analyst',
               11 : 'Electrical Engineering',
               14 : 'Health and fitness',
               19 : 'PMO',
               4: 'Business Analyst',
               9: 'DotNet Developer',
               2: 'Automation Testing',
               17: 'Netwrok Security Engineer',
               21: 'SAP Developer',
               5: 'Civil Engineer',
               0: 'Advocate',
 }

category_name = category_mapping.get(prediction_id)
print('The Resume belongs to the category of ', category_name)
```
